```python
import json
import pandas as pd

with open('dict.json','r') as f:
    dct = json.load(f)

with open('report.json','r') as f:
    report = json.load(f)
```

```python
import pandas as pd
report_df = pd.DataFrame(report)
dct_to_lst = [ {'hypothesis': k, 'score': v} for k,v in dct.items() ]
dct_df = pd.DataFrame(dct_to_lst)
```

```python
from collections import Counter
dct_df['characters'] = dct_df['hypothesis'].str.replace(" ","")
report_df = report_df[['gold']]
report_df['characters'] = report_df['gold'].str.replace(" ","")

df = report_df.merge(dct_df, on='characters', how='outer')
df = df.groupby('characters').apply(lambda x: x.sort_values('score'))
df = df.reset_index(drop=True)
df = df.assign(ord=df.groupby(['characters']).cumcount())
new_df = df[df['gold']==df['hypothesis']]
dct_counter = dict(Counter(new_df['ord'].values.tolist()))
dct_counter
```

```python
# tmp_df = new_df[new_df['ord']==1]
tmp_df = df.merge(new_df, on='characters', how='right')
mask = tmp_df['ord_x'] < 2
tmp_df = tmp_df[mask]
mask = tmp_df['ord_y'] == 1
tmp_df = tmp_df[mask]



pd.set_option('display.max_rows', len(tmp_df))
tmp_df[['hypothesis_x', 'score_x']]
```
