```python
snap_path = '../../../datasets/hashtag_segmentation/BOUN/SNAP.Hashtags.Segmented.w.Heuristics'
clean_snap_path = '../../../datasets/hashtag_segmentation/BOUN/clean.SNAP.Hashtags.Segmented.w.Heuristics'
stanford_path = '../../../datasets/hashtag_segmentation/Stanford/stanford_dataset.txt'
boun_path = '../../../datasets/hashtag_segmentation/BOUN/Test-BOUN'
```

```python
import pandas as pd

snap_df = pd.read_csv(snap_path, sep='\n', header=None)
clean_snap_df = pd.read_csv(clean_snap_path, header=None)

def stanford_reader(dataframe):
    with open(dataframe,'r') as f:
        dataset_text = f.read().split('\n')
        dataset_text = [ x.split('\t') for x in dataset_text]
        dataset_text = dataset_text[1:-1]
    data = pd.DataFrame(dataset_text)
    data = data.rename(columns={
        0: 'tweet_id',
        1: 'hashtag',
        2: 'segmentation',
        3: 'label'
    })
    data['label'] = data['label'].astype(int)
    data['hashtag'] = data['hashtag'].astype(str)
    data['hashtag'] = data['hashtag'].str.strip("'")
    data['segmentation'] = data['segmentation'].str.strip("'")
    return data

stanford_df = stanford_reader(stanford_path)
stanford_df = stanford_df[stanford_df['label']==1]

boun_df = pd.read_csv(boun_path, header=None)
```
